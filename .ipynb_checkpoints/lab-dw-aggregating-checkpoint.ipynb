```python
import pandas as pd


# Read the dataset
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv"
df = pd.read_csv(url)
df


```

```python

# Filter the dataframe
filtered_df = df[(df['Total Claim Amount'] > 1000) & (df['Response'] == 'Yes')]

# Display the filtered dataframe
print("Filtered DataFrame:")
print(filtered_df)
```

## 3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
# Group by state and count the number of customers
customer_count_by_state = df.groupby('State').size()

# Filter states with more than 500 customers
states_with_more_than_500_customers = customer_count_by_state[customer_count_by_state > 500]

# Display the filtered results
print(states_with_more_than_500_customers)
```

## 4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
# Group by education level and gender, then calculate the maximum, minimum, and median customer lifetime value
customer_lifetime_stats = df.groupby(['Education', 'Gender'])['Customer Lifetime Value'].agg(['max', 'min', 'median'])

# Display the result
print(customer_lifetime_stats)
```
